Prepare to combine both data

```python
commoncolumn_df = data_2017_df[['university','rank_display','score','country', 'year']].copy()
```

```python
pd.DataFrame(commoncolumn_df.isnull().sum(), columns=['Number of Missing values'])
```

```python
temp2017_df = commoncolumn_df.dropna()
# Check result
temp2017_df.info()
```

```python
# function to check if value can not be converted to int
def isnot_int(val):
    try:
        int_val = int(val)
        return False
    except:
        return True
```

```python
not_converted_values = temp2017_df[temp2017_df['rank_display'].map(isnot_int)]
print(len(not_converted_values),"values cannot be converted to int.")
# Check result
not_converted_values
```

```python
# Remove entries without exact ranking from data.
temp2017_df = temp2017_df.drop(not_converted_values.index)
```

```python
not_converted_values2 = temp2017_df[temp2017_df['rank_display'].map(isnot_int)]
print(len(not_converted_values2),"values cannot be converted to int.")
# Check result
not_converted_values2
```

```python
# copy selected columns from Data 2023
df_temp2023 = data_2023_df[['university','Rank','score scaled','location']].copy()
```

```python
# add year column
df_temp2023['year'] = 2023
# Check result
df_temp2023.info()
```

```python
# rename columns
df_temp2023.columns = ['university','rank_display','score','country', 'year']
# Check result
df_temp2023.info()
```
